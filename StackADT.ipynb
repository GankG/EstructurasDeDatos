```python
class StackADT:
  def __init__(self):
    self.__data = list()

  def is_empty(self):
    return len(self.__data) == 0
  
  def get_length(self):
    return len(self.__data)

  def pop(self):
    if self.is_empty():
      print("La pila est vaca y no se puede remover nada.")
    else:
      return self.__data.pop()

  def peek(self):
    if self.is_empty():
      print("Pila vaca, no se puede obtener ningn valor.")
    else:
      return self.__data[-1]
  
  def push(self, value):
    self.__data.append(value)
  
  def to_string(self):
    for item in self.__data[::-1]:
      print(f"|{item}| ")
```

```python
stack = StackADT()
print("La pila est vaca? ",stack.is_empty())
print("Cantidad de Elementos: ",stack.get_length())

stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)

print("La pila est vaca? ",stack.is_empty())
print("Cantidad de Elementos: ",stack.get_length())

stack.to_string()
stack.pop()
stack.peek()
print("\n")
stack.to_string()
```
